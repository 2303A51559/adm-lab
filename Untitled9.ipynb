```python
import pandas as pd
```

```python
data = {
 "Order_ID": [1001, 1002, 1003, 1004],
"Product": ["Laptop","Phone", "Tablet", "Laptop"],
"Price ($)": [1000, None, 500, 1200],
"Quantity": [1, 2, None, 1],
"Discount (%)": [10, 5, 0, None]
}
```

```python
df = pd.DataFrame(data)
print(df.head(2))
```

```python
print(df.isnull().sum()) # Counts missing values
```

```python
df['Price ($)'].fillna(df['Price ($)'].mean(), inplace=True)
df['Quantity'].fillna(1, inplace=True)
```

```python
df['Revenue'] = (df['Price ($)'] * df['Quantity']) - df['Discount (%)'].fillna(0)
print(df[['Order_ID', 'Revenue']])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Revenue by product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.bar(df['Product'], df['Revenue'])
plt.show()
```

```python
#draw pie chart for the given chart
import matplotlib.pyplot as plt
labels = ['Laptop', 'Phone', 'Tablet']
sizes = [1000, 500, 500]
colors = ['red', 'green', 'blue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('product distribution')
plt.show()
```

```python
#draw a graph for the given chart
import matplotlib.pyplot as plt
plt.plot(df['Order_ID'], df['Revenue'])
plt.xlabel('Order ID')
plt.ylabel('Revenue')
plt.title('Revenue Over Time')
plt.show()
```

```python
#draw box plot
import matplotlib.pyplot as plt
plt.boxplot(df['Revenue'])
plt.ylabel('Revenue')
plt.title('Box Plot of Revenue')
plt.show()
```
